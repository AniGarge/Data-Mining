```python
# Import the necessary libraries.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.linear_model import RidgeCV, LassoCV, LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
# Train data as specified in the homework instructions.

train_data=pd.read_csv("blogData_train.csv",header=None)
train_data.head()
```

```python
# Rows and columns in train data.

train_data.shape
```

```python
# Missing Values

print("Number of null values in training data:",train_data.isna().sum().sum())
```

```python
train_data.describe()
```

```python
# Reading the specified test data.

test_data=pd.read_csv("blogData_test-2012.03.31.01_00.csv",header=None)
test_data.head()
```

```python
# Rows and columns in test data.

test_data.shape
```

```python
# Separating the features and target variable in train data.

X_train=train_data.iloc[:,0:280]
y_train=train_data.iloc[:,-1]
```

```python
# Least Square Regression

linear_reg = LinearRegression()
linear_reg.fit(X_train, y_train)
```

```python
print("The intercept is:", linear_reg.intercept_)
```

```python
# Separation of features and target variable in the test data.

X_test = test_data.iloc[:,0:280]
y_test = test_data.iloc[:,-1]
```

```python
# Predictions as per the fitted linear regression model.

Y_pred=linear_reg.predict(X_test)
```

```python
print("Root mean square error for Linear Regression is:",np.sqrt(mean_squared_error(y_test,Y_pred)))
```

```python
# Ridge Regression

alphas=a_list = 10**np.linspace(-1,5, 7)
ridge_reg = RidgeCV(alphas=alphas, cv=10)                 
ridge_reg.fit(X_train, y_train)
```

```python
# Best regularization value and number of features selected.

print("Best alpha value :", ridge_reg.alpha_)
print("Features selected :", ridge_reg.coef_.nonzero()[0].size)
```

```python
# Predictions as per the fitted ridge regression model.

y_pred_ridge=ridge_reg.predict(X_test)
```

```python
print("Root mean squared error for Ridge Regression is:",np.sqrt(mean_squared_error(y_test,y_pred_ridge)))
```

```python
# Lasso Regression

lasso_reg = LassoCV(alphas=alphas, cv = 10)
lasso_reg.fit(X_train, y_train)
```

```python
# Best regularization value and features selected

print("Best alpha value :", lasso_reg.alpha_)
print("Features selected :", lasso_reg.coef_.nonzero()[0].size)
```

```python
print("Root mean squared error for Lasso Regression is:",np.sqrt(mean_squared_error(y_test,lasso_reg.predict(X_test))))
```

```python
print("Lasso picked " + str(sum(lasso_reg.coef_ != 0)) + " variables and eliminated the other " +  str(sum(lasso_reg.coef_ == 0)) + " variables")
```

```python
# Feature number and the coefficient in a descending order.

coef = pd.Series(lasso_reg.coef_, index = X_train.columns)
coef = coef.sort_values(ascending=False)
coef
```
